```python
import json
import pandas as pd
pd.get_option("display.max_columns")


with open('../results/hadoop_results_p.json') as d:
    data = json.load(d)
```

```python
df = pd.json_normalize(data)

df
```

Now that the data has been loaded, we compute the mean and the standard deviation on the ten iteration of all the combinations

```python
df = df.groupby(['falsePositiveRate', 'nReducerJob0', 'nReducerJob1', 'nLineSplitJob1']).agg(['mean', 'std'])
df = df.reset_index(['falsePositiveRate', 'nReducerJob0', 'nReducerJob1', 'nLineSplitJob1'])
df.columns = ['_'.join(col) for col in df.columns.values]
df
```

```python
df.sort_values(by=['executionTimeJob1_mean', 'executionTimeJob0_mean'])
```
